```python
import numpy as np
import gensim
import os
import re

from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim import corpora

from gensim.models.ldamulticore import LdaMulticore

import pandas as pd
```

```python
df = pd.read_csv('/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
```

```python
df.tail()
```

```python
STOPWORDS = set(STOPWORDS).union(set(['said', 'mr', 'mrs']))

def tokenize(text):
    return [token for token in simple_preprocess(text) if token not in STOPWORDS]
```

```python
def doc_stream(path):
  for f in path:
    text = f
    tokens = tokenize(str(text))
    yield tokens
```

```python
test = doc_stream(df['reviews.text'])
```

```python
id2word = corpora.Dictionary(doc_stream(df['reviews.text']))
```

```python
id2word.filter_extremes(no_below=20, no_above=0.85)
```

```python
corpus = [id2word.doc2bow(text) for text in doc_stream(df['reviews.text'])]
```

```python
lda = LdaMulticore(corpus=corpus,
                   id2word=id2word,
                   random_state=47,
                   num_topics = 15,
                   passes=10,
                   workers=8
                  )
```

```python
lda.print_topics()
```

```python
words = [re.findall(r'"([^"]*)"',t[1]) for t in lda.print_topics()]
```

```python
topics = [' '.join(t[0:5]) for t in words]
```

```python
for t in topics: 
    print(t)
    print("\n")
```
